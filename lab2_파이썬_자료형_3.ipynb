## 딕셔너리 자료형

*  딕셔너리 자료형: key와 value를 한 쌍으로 갖는 자료형
*  my_dic = {Key1: value1, Key2: value2, key3: value3 … }


```python
my_dic = {'name': 'pey', 'phone': '0119993323', 'birth': '1118'}

print (my_dic['name'])
print (my_dic['phone'])
print (my_dic['birth'])
```

* 딕셔너리의 value로 리스트를 넣을 수 있고, 딕셔너리의 key로 튜플을 넣을 수 있음
* 딕셔너리의 value로 또 다른 딕셔너리가 들어갈 수 있음


```python
a = {('a', 1): 'b'}
print (a)

a = {'a':[1,2,3]}
print (a)

a = {'a':{'b': 1}}
print (a)
print (a['a']['b'])
```

딕셔너리 쌍 추가하기, 요소 제거하기

```python
a = {}
a[1] = 'a'
a[2] = 'b'
print (a)

del a[1]
print (a)
```

딕셔너리에서 key 사용하여 value 얻기

```python
grade = {'pey': 10}
print (grade['pey'])

```

딕셔너리 만들 때 주의사항

```python
a = {1:'a', 1:'b'}
print (a)
```

Key, value 쌍을 얻기(items)

```python
a = {'name':'pey', 'phone':'0119993323', 'birth':'1118'}
print (a.items())
```

Key, value 쌍 모두 지우기(clear)

```python
a = {'name':'pey', 'phone':'0119993323', 'birth':'1118'}
print (a.items())
a.clear()
print (a)
```

Key로 value 얻기

```python
a = {'name':'pey', 'phone':'0119993323', 'birth':'1118'}
print(a.get('name'))
print(a['name']) # 같은 문법
```

해당 key가 딕셔너리 안에 있는지 조사하기 (in)

```python
a = {'name':'pey', 'phone':'0119993323', 'birth':'1118'}
print('name' in a)
print('email' in a)
```
