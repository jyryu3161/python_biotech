### 이스케이프 코드 (Escape Codes)

escape code.png

"이스케이프 코드 (escape code)"문자열을 잠시 탈출해서 다른 할일을 하고 다시 문자열로 돌아온다.

```python
multiline = "Life is too short.\n You need python."
print (multiline)

test_escape_code_line = "Life is too short.\t You need python."
print (test_escape_code_line)

test_escape_code_line = "\"Life is too short.\t You need python.\""
print (test_escape_code_line)
```
