## 반복문 (while 문)

While 문의 <조건문>이 True면 들여쓰기 한 부분이 수행됨

```python
while <조건문>: 
    <수행할 문장1>
    <수행할 문장2>
    <수행할 문장3>
```

```python
treeHit = 0
while treeHit < 10: # threeHit이 10 미만인동안 수행
    treeHit = treeHit + 1 # treeHit += 1
    print ("나무를 %d번 찍었습니다."%(treeHit))
    if treeHit == 10:
        print ("나무 넘어갑니다.")

        
```

break: 즉시 반복문 밖으로 나감. 들여써진 부분 밖으로

```python
coffee = 10
money = 300
while money: # money가 True인 동안 수행
    print ("돈을 받았으니 커피를 줍니다.")
    coffee = coffee - 1
    print ("남은 커피의 양은 %d개 입니다."%(coffee))
    if coffee == 0: # if not coffee:
        print ("커피가 다 떨어졌습니다. 판매를 중지합니다.")
        break
```

continue: 반복문의 시작점으로 이동

```python
a = 0
while a < 10:
    a = a + 1
    if a % 2 == 0:
        continue
    print (a)

    
```

## for 문의 기본구조

```python
for 변수 in 리스트(또는 튜플, 문자열):
    수행할 문장1
    수행할 문장2
    ...
```

전형적인 for 문

```python
test_list = ['one', 'two', 'three']
for i in test_list:
    print (i)
print("")

test_list = ('one', 'two', 'three')
for i in test_list:
    print (i)
print("")

test_list = 'one'
for i in test_list:
    print (i)
```

다양한 for 문의 사용

```python
a = [(1, 2), (3, 4), (5, 6)]
for (first, last) in a:
    print(first + last)
    
```

60점이 넘으면 합격이고 그렇지 않으면 불합격

```python
marks = [90, 25, 67, 45, 80]

number = 0
for mark in marks:
    number = number + 1
    if mark >= 60:
        print("%d번 학생은 합격입니다."%(number))
    else:
        print("%d번 학생은 불합격입니다."%(number))
        
```

for문과 continue

```python
marks = [90, 25, 67, 45, 80]

number = 0
for mark in marks:
    number = number + 1
    if mark < 60:
        continue
    
    print("%d번 학생 축하합니다. 합격입니다."%(number))
        
```

반복문의 중첩

```python
for i in range(2, 10):
    for j in range(1, 10):
        print(i*j, end=" ")
    print(" ")
    
```

리스트 내포(List comprehension)

```python
result = [num * 3 for num in range(1, 5)]
print (result)

result = []
for num in range(1, 5):
    result.append(num*3)
print (result)
```
